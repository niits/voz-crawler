```python
URL = "https://voz.vn/t/thread-tong-hop-chia-se-ve-muc-luong-tai-cac-cong-ty-part-2.515355"
```

```python
from voz_crawler import VozCrawler

crawler = VozCrawler(delay=1.0)

# Crawl trang hiện tại
page_data = crawler.crawl_page(URL)
print(f"Thread: {page_data.thread_url}")
print(f"Page {page_data.current_page}/{page_data.total_pages}")
print(f"Số bài viết: {len(page_data.posts)}")

# Chuyển sang DataFrame
df = VozCrawler.pages_to_dataframe([page_data])
df[["username", "datetime", "content_text", "reaction_count"]].head(20)
```

```python
from tqdm.auto import tqdm

total_pages = page_data.total_pages
all_pages = [page_data]  # đã có trang 1

for p in tqdm(range(2, total_pages + 1), desc="Crawling", initial=1, total=total_pages):
    try:
        pd_data = crawler.crawl_page(URL, page=p)
        all_pages.append(pd_data)
    except Exception as e:
        print(f"\n⚠ Page {p} failed: {e}")

print(f"\nDone! Crawled {len(all_pages)}/{total_pages} pages")

df_all = VozCrawler.pages_to_dataframe(all_pages)
print(f"Total posts: {len(df_all)}")
df_all.head(10)
```

```python
df.to_json("../data/voz_salaries_part2.json", orient="records", lines=True)
```
